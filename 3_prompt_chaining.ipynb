```python
from langgraph.graph import StateGraph, START, END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
# Create state
class BlogState(TypedDict) :
    title: str
    outline: str
    content:str
    evaluate: str
```

```python
def create_outline(state:BlogState)->BlogState:
    # fetch title
    title = state['title']

    #create prompt
    prompt = f'Generate an outline for the a blog on the topic {title}'

    # call llm
    outline = model.invoke(prompt).content
    state['outline'] = outline

    return state
```

```python
def create_blog(state:BlogState)->BlogState:
    title = state['title']
    outline = state['outline']
    prompt = f'Generate a detailed blog on the topic {title} using the follwing outline \n {outline }'
    content = model.invoke(prompt).content

    state['content'] = content
    return state
```

```python
def evaluate(state:BlogState)->BlogState:
    title = state['title']
    outline = state['outline']
    content = state['content']
    prompt = f'Based on the topic {title} With the follwing outline: \n {outline} \n and the content of the blog  \n {content} \n  Evaluate and score the blog'
    evaluate = model.invoke(prompt).content

    state['evaluate'] = evaluate
    return state
```

```python
graph = StateGraph(BlogState)
# add nodes
graph.add_node('create_outline' , create_outline)
graph.add_node('create_blog' , create_blog)
graph.add_node('evaluate' , evaluate)


# add edges
graph.add_edge(START,'create_outline' )
graph.add_edge('create_outline', 'create_blog')
graph.add_edge('create_blog','evaluate')
graph.add_edge('evaluate',END)



# compile
workflow = graph.compile()
```

```python
initial_state = {'title':'Rise of AI in Pakistan'}
final_state = workflow.invoke(initial_state)
```
